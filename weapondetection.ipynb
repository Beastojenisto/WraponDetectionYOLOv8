# **Creating The Config file in the current directory**

```python
import yaml

data_yaml = dict(
    train ='/kaggle/working/Weapon-Detection-2//train',
    val ='/kaggle/working/Weapon-Detection-2//valid',
    test='/kaggle/working/Weapon-Detection-2//test',
    nc = 3,
    names =['gun','guns','handgun']
)

with open('data.yaml', 'w') as outfile:
    yaml.dump(data_yaml, outfile, default_flow_style=True)
    
%cat data.yaml
```

# **Training the YOLOv8 large model**

```python
from ultralytics import YOLO
model = YOLO('yolov8l.pt')
model.train(data='/kaggle/working/data.yaml', epochs=50, imgsz=416, batch=16,exist_ok=True)
```
